# Data visualization

## Average price by the number of beds

```python
average_prices = df.groupby('beds')['price'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(average_prices['beds'], average_prices['price'], color='skyblue')
plt.xlabel('Number of beds')
plt.ylabel('Average Price')
plt.title('Average Price vs. Number of Beds')
plt.show()
```

```python
average_prices = df.groupby('bedrooms')['price'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(average_prices['bedrooms'], average_prices['price'], color='skyblue')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Average Price')
plt.title('Average Price vs. Number of Bedrooms')
plt.show()
```

```python
average_prices_by_neighborhood = df.groupby('neighbourhood_cleansed')['price'].mean().reset_index()

average_prices_by_neighborhood = average_prices_by_neighborhood.sort_values(by='price', ascending=True)

plt.figure(figsize=(12, 8))
plt.barh(average_prices_by_neighborhood['neighbourhood_cleansed'], average_prices_by_neighborhood['price'], color='skyblue')
plt.xlabel('Average Price')
plt.ylabel('Region name')
plt.title('Average Price by Location')
plt.tight_layout()
plt.show()
```

# Feature Engineering

In this dataset, there was one thing that exptremely got my interest. In the amenities column, there is a list of amenities per listings, So I wanted to see the price correlation of whether or not the particular amenity exists in the listing. For that I am going to use the 'coffe maker' amenity as the new variable for my analysis.

```python
df.info()
```

Lets see their counts first

```python
average_prices_by_feature = df.groupby('HasCoffeeMaker')['id'].count()

average_prices_by_feature.plot(kind='bar')
plt.xlabel('Coffe Maker')
plt.ylabel('Count of Listings')
plt.title('Count of Listings that have and do not have a Coffee Maker')
plt.xticks(rotation=0)
plt.show()
```

```python
average_prices_by_feature = df.groupby('HasCoffeeMaker')['price'].mean()

average_prices_by_feature.plot(kind='bar')
plt.xlabel('Coffe Maker')
plt.ylabel('Average Price')
plt.title('Average Price by Wheather or not the listing has a Coffee Maker')
plt.xticks(rotation=0)
plt.show()
```

# Conclusion

### The effect of number of Bedrooms and number of Beds:
As predicted, apparently the dependency between the number of Bedrooms & number of Beds and listing price is linear. The more bedroom and the beds the listing has, the more money it cost (Obviously).

### Location effect:
From the analysis, we can say that, "Indre By" has the most expencive average price while "Brnshj-Husum" has the cheapest average price among all of the regions of the Copenhagen.

### Amenity effect:
From the analysis we can see that there are more listing that doesn't include Coffee Maker than vice versa. And the average price of the listings with Coffee Maker are slightly higher.
